```python
import pandas as pd
```

```python
spx = pd.read_csv('SPX_index_sample/SPX_1min_sample.csv')
escont = pd.read_csv('ES_futures_sample/ES_continuous_1min_sample.csv')
```

```python
from matplotlib import pyplot as plt
import numpy as np
import datetime as datetime

pd.to_datetime(spx.timestamp, format="%Y-%m-%d %H:%M:%S")
pd.to_datetime(escont.timestamp, format="%Y-%m-%d %H:%M:%S")
spx['timestamp'] = spx.timestamp.map(lambda datestring: np.datetime64(datestring) - np.datetime64("2022-04-01 00:00:00")).map(lambda ts: int(ts / np.timedelta64("1000000000", "ns")))
escont['timestamp'] = escont.timestamp.map(lambda datestring: np.datetime64(datestring) - np.datetime64("2022-04-01 00:00:00")).map(lambda ts: int(ts / np.timedelta64("1000000000", "ns")))

```

```python
plt.plot(spx['timestamp'], spx['open'], label="SPX")
plt.plot(escont['timestamp'], escont['open'], label="ES Continuous Futures")
plt.savefig("test_image1.png", dpi=500)
```
